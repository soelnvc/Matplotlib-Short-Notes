## 📊 Step 1: Bar Charts (Categorical Comparisons)

Bar charts are used to compare quantities across different categories. In this example, we visualize smartphone market share.

### 🛠️ Key Functions & Arguments

#### 1. `plt.bar(x, height, ...)`
This is the primary function for vertical bar charts.
* **`x`**: The names of the categories (e.g., Brands).
* **`height`**: The numerical values for each category.
* **`color`**: Can be a single color or a list of colors for each bar (as seen in your code).
* **`width`**: (Optional) Changes the thickness of the bars (default is 0.8).
* **`edgecolor`**: The color of the border around the bars.
* **`linewidth`**: Thickness of the bar borders.

#### 2. `plt.barh()`
* Want horizontal bars? Just swap `bar` for `barh`. Note that you use `width` for data values and `height` for bar thickness here.


---
### 💡 Pro-Tip: Adding Value Labels
A common professional touch is adding the exact value on top of each bar using `plt.text()` or the modern `ax.bar_label()` method.

````python
# Bar Chart
categories = np.array(['Samsung', 'Apple', 'Huawei', 'Xiaomi', 'Oppo']) # Numpy arrays are not fater for string data but I used it cause I love them LOL😂
values = np.array([30, 25, 20, 15, 10])

plt.bar(categories, values, color=['#2C3E50', '#E74C3C', '#3498DB', '#9B59B6', '#F39C12']) # Custom colors for each bar
# plt.barh() for horizontal bar chart
plt.xlabel('Brands', fontsize=12, fontweight='bold', color='#34495E')
plt.ylabel('Market Share (%)', fontsize=12, fontweight='bold', color='#34495E')
plt.title('Smartphone Market Share', fontsize=14, fontweight='bold', color='#2C3E50')
plt.grid(axis='y', linestyle='--', alpha=0.7) # Grid only on y-axis for better readability
plt.show()
````

## 🍕 Step 3: Pie Charts (Proportions)

Pie charts are ideal for showing how a total amount is divided into parts. Your example uses programming language popularity to demonstrate this.

### 🛠️ Key Functions & Arguments

#### 1. `plt.pie(x, ...)`
* **`labels`**: A list of strings to label each "slice."
* **`colors`**: A list of colors to apply to the slices.
* **`autopct`**: (Super Important) Formats the value labels inside the slices. 
    * Example: `'%1.1f%%'` shows one decimal place followed by a % sign.
* **`explode`**: An array that "offsets" a slice from the center (perfect for highlighting the most important data point).
* **`startangle`**: Rotates the start of the chart (e.g., `90` or `140`).
* **`shadow`**: Set to `True` for a 3D-depth effect.
* **`pctdistance`**: The distance of the percentage text from the center (default is 0.6).

#### 2. `plt.axis('equal')`
* **Crucial:** This ensures that the pie chart is rendered as a perfect circle rather than an oval.

````python
#Pie Chart
labels = ['Python', 'JavaScript', 'Java', 'C#', 'PHP']
sizes = [30, 25, 20, 15, 10]
colors = ['#2C3E50', '#E74C3C', '#3498DB', '#9B59B6', '#F39C12']
explode = (0.05, 0, 0, 0, 0) # Explode the first slice (Python) for emphasis
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.2f%%', startangle=140, shadow=True) #Keywords: autopct for percentage display, startangle for rotation
plt.title('Programming Language Popularity', fontsize=14, fontweight='bold', color='#2C3E50')
plt.axis('equal') # Equal aspect ratio ensures that pie chart is circular
plt.show()
````

## 📍 Step 4: Scatter Plots (Relationships)

Scatter plots are used to observe the relationship (correlation) between two numerical variables.

### 🛠️ Key Functions & Arguments

#### 1. `plt.scatter(x, y, ...)`
* **`s`**: Changes the **size** of the points.
* **`c`**: Changes the **color**. You can even pass an array of values here to create a "heatmap" effect.
* **`marker`**: Change the shape of points (e.g., `'x'`, `'s'`, `'v'`).
* **`alpha`**: Transparency. Very useful for "Overplotting" (when points overlap, higher transparency helps you see the density).
* **`cmap`**: If using a color array, this defines the color theme (e.g., `'viridis'`, `'magma'`).

#### 2. `plt.legend()`
* **`loc`**: Placement (e.g., `'upper right'`, `'lower left'`).
* **`frameon`**: `True/False` to show the box around the legend.

````python
# Scatter Plot
x1 = np.random.rand(100)
y1 = np.random.rand(100) # I am using random data for demonstration, cause I ran out of Examples LOL XD, you can replace it with your actual data
plt.scatter(x1, y1, color='#3498DB', alpha=0.7, s=50) # alpha controls the transparency of the points, s controls the size of the points
x2 = np.random.rand(100)
y2 = np.random.rand(100)
plt.scatter(x2, y2, color='#E74C3C', alpha=0.7, s=50) # Adding a second set of points for comparison
plt.xlabel('X-axis', fontsize=12, fontweight='bold', color='#34495E')
plt.ylabel('Y-axis', fontsize=12, fontweight='bold', color='#34495E')
plt.title('Random Scatter Plot', fontsize=14, fontweight='bold', color='#2C3E50')
plt.grid(True, which='both', linestyle='--', linewidth=0.5, alpha=0.7) # Grid for better visibility of points
plt.legend(['Dataset 1', 'Dataset 2'], loc='upper right') # Legend to differentiate between the two datasets    
plt.show()
````

## 📈 Step 5: Histograms (Data Density)

Histograms show the distribution of a continuous variable by "binning" the data.

### 🛠️ Key Functions & Arguments

#### 1. `plt.hist(x, ...)`
* **`bins`**: The number of intervals to group data into. More bins = more detail.
* **`range`**: The lower and upper range of the bins.
* **`density`**: If `True`, the first element of the return tuple will be the counts normalized to form a probability density.
* **`cumulative`**: If `True`, a histogram is computed where each bin gives the counts plus all bins for smaller values.
* **`histtype`**: Type of histogram to draw (e.g., `'bar'`, `'step'`, `'stepfilled'`).
* **`orientation`**: `'vertical'` or `'horizontal'`.

#### 2. `np.random.normal(loc, scale, size)`
* In your code, you used this to generate a "Bell Curve."
* **`loc`**: The mean (center) of the distribution.
* **`scale`**: The standard deviation (spread).

````python
# Histogram
data = np.random.normal(loc=0, scale=0.2, size=1000) # Generating random data from a normal distribution
plt.hist(data, bins=100, color="#B4FF85", edgecolor="#1E4D00", alpha=0.7) # bins for number of bars, edgecolor for bar borders
plt.xlabel('Value', fontsize=12, fontweight='bold', color='#34495E')
plt.ylabel('Frequency', fontsize=12, fontweight='bold', color='#34495E')
plt.title('Random Data Distribution', fontsize=14, fontweight='bold', color='#2C3E50')
plt.grid(axis='y', linestyle='--', alpha=0.7) # Grid only on y-axis for better readability
plt.show()
````
